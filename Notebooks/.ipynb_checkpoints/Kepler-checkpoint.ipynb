### Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost import XGBRegressor, XGBClassifier, plot_tree
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import auc, accuracy_score, confusion_matrix, mean_squared_error
from sklearn import tree

df = pd.read_csv('../Data/kepler.csv')
```

### EDA

```python
confirmed = df[df['koi_disposition']=='CONFIRMED']
candidate = df[df['koi_disposition']=='CANDIDATE']
fp = df[df['koi_disposition']=='FALSE POSITIVE']
```

```python
# plt.hist(fp['koi_srad']);
```

### Dropped empty rows for chosen columns

```python
df_filtered = df.dropna(subset=['koi_depth','koi_prad','koi_teq'])
```

### Replaced dispositions with numeric values for XGBoost to work

```python
df_filtered['koi_disposition'].replace('FALSE POSITIVE',0,inplace=True)
df_filtered['koi_disposition'].replace('CANDIDATE',1,inplace=True)
df_filtered['koi_disposition'].replace('CONFIRMED',2,inplace=True)
```

```python
df_filtered['koi_disposition'].value_counts()
```

### Set our X and y

```python
y = df_filtered['koi_disposition']
X = df_filtered[['koi_depth','koi_prad','koi_teq']]
```

### Running train_test_split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25)
train = xgb.DMatrix(X_train,label=y_train)
test = xgb.DMatrix(X_test,label=y_test)
```

### Changing parameters to see how they affect resulting confusion matrix

```python
xgb_model = xgb.XGBClassifier(booster = 'gbtree', eta = 0.25, max_depth = 5, objective="multi:softmax", random_state=42)
xgb_model.fit(X_train, y_train)
print("Training set accuracy score:",xgb_model.score(X_train, y_train))
print("Test set accuracy score:",xgb_model.score(X_test, y_test))
y_pred = xgb_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
```

### Plot Tree

```python
plot_tree(xgb_model, rankdir="LR");
fig = plt.gcf()
fig.set_size_inches(100,100)
```
